## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.xlabel('indices')
plt.ylabel('Numbers')
plt.title('myplot')
plt.show()    # By default x values are [0,1,2,3] as the numbers are not specified
```

```python
plt.plot([1,2,3,4] , [1,4,9,16])
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Square of corresponding number')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4] , [1,4,9,16] , 'ro')
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Square of corresponding number')
plt.grid()
plt.show()
```

```python
t = np.arange(0 , 5. , 0.2)
# blue dashes , red squares , green triangles
plt.plot(t , t**2 , 'b--' , label = '^2')
plt.plot(t , t**2.2 , 'rs' , label = '^2.2')
plt.plot(t , t**2.5 , 'g^' , label = '^2.5')
plt.legend()
plt.grid()
plt.show()
```

## Controlling line properties

```python
plt.plot([1,2,3,4] , [1,4,9,16] , linewidth = 5.0)
plt.grid()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0] , color = 'r' , linewidth = '2.0')     #python style of giving arguments
plt.setp(lines[1] , 'color' , 'g' , 'linewidth' , '2.0') #matlab style of giving arguments
plt.grid()
plt.show()
```

## Working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0. , 5. , 0.1)
t2 = np.arange(0. , 5. , 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1 , f(t1) , 'b-')
plt.grid()

plt.subplot(212)
plt.plot(t2 , np.cos(2*np.pi*t2) , 'r--')
plt.grid()

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3] , 'r')
plt.subplot(212)
plt.plot([4,5,6] , 'g')


plt.figure(2)

plt.plot([1,2,3] , 'r')

plt.figure(1)
plt.subplot(211)
plt.title('This is figure 1')

plt.show()
```
