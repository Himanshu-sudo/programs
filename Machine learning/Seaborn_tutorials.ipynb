# Seaborn

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(x = 'day' , y = 'tip' , data = tips)
```

```python
sns.barplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
sns.barplot(x = 'day' , y = 'tip' , data = tips , hue = 'sex') # By default estimator is mean
```

```python
sns.barplot(x = 'total_bill' , y = 'day' , data = tips)
# sns.barplot(x = 'total_bill' , y = 'day' , data = tips , palette = 'winter')
```

```python
sns.barplot(x = 'day' , y = 'tip' , data = tips , order = ['Sat' , 'Fri' , 'Sun' , 'Thur'])
```

```python
from numpy import median
```

```python
sns.barplot(x = 'day' , y = 'tip' , data = tips , estimator = median)
```

```python
sns.barplot(x = 'day' , y = 'tip' , data = tips , ci = 34 , capsize=0.3) #smaller error bars
```
